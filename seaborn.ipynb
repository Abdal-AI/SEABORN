```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=sns.load_dataset("penguins")



sns.lineplot(data=x,x="bill_length_mm",y="flipper_length_mm",hue="species",style="species",markers=["o",">","*"],palette="Accent")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset and get the first 10 rows
df = pd.read_csv("Breast_cancer_dataset.csv").head(10)

# Create the line plot with corrected column names
sns.lineplot(data=df, x="radius_mean", y="texture_mean", hue="diagnosis", style="diagnosis", markers=["o", "*"], palette="Accent")

# Add labels and title for clarity
plt.title("Line Plot of Mean Texture vs Mean Radius")
plt.xlabel("Mean Radius")
plt.ylabel("Mean Texture")

plt.show()
```
